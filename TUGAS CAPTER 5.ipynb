```python
import pandas as pd                                     # Mengimpor pustaka pandas untuk manipulasi dan analisis data tabular (seperti Excel/CSV)
from scipy.stats.mstats import gmean, hmean             # Mengimpor fungsi rata-rata geometrik (gmean) dan harmonik (hmean) dari pustaka SciPy
import matplotlib.pyplot as plt                         # Mengimpor pyplot dari matplotlib untuk membuat visualisasi data/grafik

# Membaca file Excel bernama 'Smith_glass_post_NYT_data.xlsx', khusus pada sheet yang bernama 'Supp_traces'
my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

a_mean = my_dataset.Sr.mean()                           # Menghitung rata-rata aritmetika biasa dari kolom 'Sr' (Strontium)
g_mean = gmean(my_dataset['Sr'])                        # Menghitung rata-rata geometrik dari kolom 'Sr' menggunakan SciPy
h_mean = hmean(my_dataset['Sr'])                        # Menghitung rata-rata harmonik dari kolom 'Sr' menggunakan SciPy

print ('-------')                                       # Mencetak garis pemisah
print ('arithmetic mean')                               # Mencetak teks judul untuk rata-rata aritmetika
print ("{0:.1f} [ppm]".format(a_mean))                  # Mencetak hasil a_mean dengan format 1 angka di belakang koma beserta satuannya [ppm]
print ('-------')                                       # Mencetak garis pemisah

print ('geometric mean')                                # Mencetak teks judul untuk rata-rata geometrik
print ("{0:.1f} [ppm]".format(g_mean))                  # Mencetak hasil g_mean dengan format 1 angka di belakang koma
print ('-------')                                       # Mencetak garis pemisah

print ('harmonic mean')                                 # Mencetak teks judul untuk rata-rata harmonik
print ("{0:.1f} [ppm]".format(h_mean))                  # Mencetak hasil h_mean dengan format 1 angka di belakang koma
print ('-------')                                       # Mencetak garis pemisah

fig, ax = plt.subplots()                                # Menginisialisasi figure (kanvas) dan ax (sumbu) untuk tempat menggambar grafik
# Membuat histogram dari kolom 'Sr'. Bins ditentukan otomatis, density=True membuat sumbu Y menjadi probabilitas (bukan frekuensi hitungan mutlak), edgecolor='k' memberi garis tepi hitam pada batang, dan alpha mengatur transparansi
ax.hist(my_dataset.Sr, bins='auto', density=True, edgecolor='k', label='Measurements Hist', alpha=0.8) 
# Menambahkan garis vertikal (axvline) pada posisi rata-rata aritmetika, berwarna ungu
ax.axvline(a_mean, color='purple', label='Arithmetic mean', linewidth=2)
# Menambahkan garis vertikal pada posisi rata-rata geometrik, berwarna hijau
ax.axvline(g_mean, color='green',  label='Geometric mean', linewidth=2)
# Menambahkan garis vertikal pada posisi rata-rata harmonik, berwarna pink
ax.axvline(h_mean, color='pink',  label='Harmonic mean', linewidth=2)
ax.set_xlabel('Sr [ppm]')                               # Memberi nama label pada sumbu X
ax.set_ylabel('Probability density')                    # Memberi nama label pada sumbu Y
ax.legend()                                             # Menampilkan legenda/keterangan grafik yang mengambil dari argumen 'label' di atas

'''
Output:
-------
arithmetic mean
365.4 [ppm]
-------
geometric mean
348.6 [ppm]
-------
harmonic mean
333.8 [ppm]
-------
'''
```

```python
import pandas as pd                                     # Mengimpor pandas untuk manipulasi data (membaca file Excel)
import numpy as np                                      # Mengimpor numpy untuk komputasi numerik matematis
import matplotlib.pyplot as plt                         # Mengimpor matplotlib untuk membuat visualisasi grafik

# Membaca data dari file Excel 'Smith_glass_post_NYT_data.xlsx' pada sheet 'Supp_traces'
my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

# Menghitung histogram secara matematis menggunakan NumPy. 
# 'hist' menyimpan nilai kepadatan (probabilitas) tiap bin, 'bin_edges' menyimpan angka batas-batas tiap bin
hist, bin_edges = np.histogram(my_dataset['Pr'], bins= 10, density=True)

# Menghitung estimasi nilai modus (modal value). 
# hist.argmax() mencari indeks bin dengan puncak tertinggi. Lalu kita ambil titik tengah dari bin tersebut dengan menjumlahkan batas kiri dan kanan lalu dibagi 2
modal_value = (bin_edges[hist.argmax()] +  bin_edges[hist.argmax()+1])/2

# Mencetak hasil perhitungan modus tanpa angka desimal di belakang koma ({0:.0f}) beserta satuannya
print ('modal value: {0:.0f} [ppm]'.format(modal_value))

fig, ax = plt.subplots()                                # Menyiapkan kanvas (fig) dan sumbu grafik (ax)
# Menggambar histogram dari kolom 'Pr' (Praseodymium) dengan 10 kelompok (bins), garis tepi hitam, dan transparansi 80%
ax.hist(my_dataset.Pr, bins=10, density=True, edgecolor='k', label="Measurements Hist", alpha=0.8) 
# Menambahkan garis vertikal pada titik nilai modus, berwarna oranye dengan ketebalan 2
ax.axvline(modal_value, color="orange", label="Modal value", linewidth=2)
ax.set_xlabel('Pr [ppm]')                               # Memberi nama label pada sumbu X
ax.set_ylabel('Probability density')                    # Memberi nama label pada sumbu Y
ax.legend()                                             # Memunculkan kotak legenda di dalam grafik

'''
Output: modal value: 277 [ppm]
'''
```

```python
import pandas as pd                                     # Mengimpor pustaka pandas untuk membaca dan memanipulasi data
import matplotlib.pyplot as plt                         # Mengimpor pustaka matplotlib untuk membuat visualisasi grafik

# Membaca data dari file Excel 'Smith_glass_post_NYT_data.xlsx' pada sheet 'Supp_traces'
my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

variance =  my_dataset['Nd'].var()                      # Menghitung nilai varians (ragam) dari kolom 'Nd' (Neodymium)
stddev =  my_dataset['Nd'].std()                        # Menghitung standar deviasi (simpangan baku) dari kolom 'Nd'

print ('-------')                                       # Mencetak garis pemisah
print ('Variance')                                      # Mencetak teks "Variance"
print("{0:.0f} [square ppm]".format(variance))          # Mencetak nilai varians, dibulatkan tanpa angka desimal dengan satuan [square ppm]
print ('-------')                                       # Mencetak garis pemisah
print ('Standard Deviation')                            # Mencetak teks "Standard Deviation"
print("{0:.0f} [ppm]".format(stddev))                   # Mencetak nilai standar deviasi, dibulatkan tanpa angka desimal dengan satuan [ppm]
print ('-------')                                       # Mencetak garis pemisah penutup

fig, ax = plt.subplots()                                # Menyiapkan kanvas (fig) dan area sumbu (ax) untuk menggambar grafik
# Membuat histogram dari kolom 'Nd' dengan 10 kelas (bins), sumbu Y berupa probabilitas (density=True), dan batas batang berwarna hitam
ax.hist(my_dataset.Nd, bins= 10, density = True, edgecolor='k', label='Measurements Hist') 
# Menambahkan garis vertikal penanda batas bawah (rata-rata dikurangi 1 standar deviasi), berwarna ungu
ax.axvline(my_dataset['Nd'].mean()-stddev, color='purple', label=r'mean - 1$\sigma$', linewidth=2)
# Menambahkan garis vertikal penanda batas atas (rata-rata ditambah 1 standar deviasi), berwarna pink
ax.axvline(my_dataset['Nd'].mean()+stddev, color='pink', label=r'mean + 1$\sigma$', linewidth=2)
# Membuat area berbayang (span) di antara batas bawah dan batas atas dengan transparansi 10% (alpha=0.1) untuk menyoroti sebaran 1 standar deviasi
ax.axvspan(my_dataset['Nd'].mean()-stddev, my_dataset['Nd'].mean()+stddev, alpha=0.1, color='pink', label=r'mean $\pm$ 1$\sigma$')
ax.set_xlabel('Nd [ppm]')                               # Memberi nama label pada sumbu X
ax.set_ylabel('Probability density')                    # Memberi nama label pada sumbu Y
ax.legend()                                             # Memunculkan kotak keterangan (legenda) grafik

'''
Output:
-------
Variance
14021 [square ppm]
-------
Standard Deviation
118 [ppm]
-------
'''
```

```python
import pandas as pd                                     # Mengimpor pustaka pandas untuk membaca dan memanipulasi data tabular
import numpy as np                                      # Mengimpor pustaka numpy untuk komputasi numerik (menghitung persentil)
import matplotlib.pyplot as plt                         # Mengimpor pustaka matplotlib untuk membuat visualisasi grafik

# Membaca data dari file Excel 'Smith_glass_post_NYT_data.xlsx' spesifik pada sheet 'Supp_traces'
my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

# Menghitung Kuartil 1 (Q1) atau persentil ke-25 dari kolom 'U' (Uranium) menggunakan metode titik tengah (midpoint)
q1 = np.percentile(my_dataset.U, 25,method = 'midpoint') 
# Menghitung Kuartil 3 (Q3) atau persentil ke-75 dari kolom 'U' menggunakan metode titik tengah
q3 = np.percentile(my_dataset.U, 75,method = 'midpoint') 
  
iqr = q3 - q1                                           # Menghitung Interquartile Range (IQR) dengan mengurangkan Q3 dengan Q1  

print ('-------')                                       # Mencetak garis pemisah
# Mencetak nilai IQR yang dibulatkan tanpa angka desimal ({0:.0f}) beserta satuannya [ppm]
print ('Interquaritle range (IQR): {0:.0f} [ppm]'.format(iqr))
print ('-------')                                       # Mencetak garis pemisah penutup

fig, ax = plt.subplots()                                # Menyiapkan kanvas (fig) dan area sumbu (ax) untuk menggambar grafik
# Membuat histogram dari kolom 'U', penentuan kelompok (bins) otomatis, sumbu Y sebagai probabilitas (density), dan batas batang hitam
ax.hist(my_dataset.U, bins='auto', density=True, edgecolor='k', label='Measurements Hist') 
# Menambahkan garis vertikal penanda Kuartil 1 (Q1), berwarna ungu dengan ketebalan 2
ax.axvline(q1, color='purple', label='Q1', linewidth=2)
# Menambahkan garis vertikal penanda Kuartil 3 (Q3), berwarna pink dengan ketebalan 2
ax.axvline(q3, color='pink', label='Q3', linewidth=2)
# Membuat area berbayang (span) di antara Q1 dan Q3 dengan transparansi 10% (alpha=0.1) untuk menyoroti area sebaran IQR
ax.axvspan(q1, q3, alpha=0.1, color='pink', label='Interquaritle range (IQR)')
ax.set_xlabel('U [ppm]')                                # Memberi nama label pada sumbu X
ax.set_ylabel('Probability density')                    # Memberi nama label pada sumbu Y
ax.legend()                                             # Memunculkan kotak keterangan (legenda) grafik

'''
Output:
-------
Interquaritle range (IQR): 164 [ppm]
-------
'''
```

```python
import pandas as pd                                     # Mengimpor pustaka pandas untuk membaca dan memanipulasi data tabular
import numpy as np                                      # Mengimpor pustaka numpy untuk komputasi numerik matematis
import matplotlib.pyplot as plt                         # Mengimpor pustaka matplotlib untuk membuat visualisasi grafik

# Membaca data dari file Excel 'Smith_glass_post_NYT_data.xlsx' pada sheet 'Supp_traces'
my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

a_mean = my_dataset.Th.mean()                           # Menghitung nilai rata-rata aritmetika (mean) dari kolom 'Th' (Thorium)

median = my_dataset.Th.median()                         # Menghitung nilai tengah (median) dari kolom 'Th'

# Menghitung histogram dengan 20 kelompok (bins) menggunakan NumPy untuk keperluan mencari modus
hist, bin_edges = np.histogram(my_dataset['Th'], bins=20, density=True)
# Mengestimasi nilai modus dengan mencari bin yang memiliki puncak (probabilitas) tertinggi, lalu mengambil nilai tengah dari bin tersebut
modal_value = (bin_edges[hist.argmax()] +  bin_edges[hist.argmax()+1])/2

fig, ax = plt.subplots()                                # Menyiapkan kanvas (fig) dan area sumbu (ax) untuk menggambar grafik
# Membuat histogram visual dari kolom 'Th'. Catatan: di sini Anda menggunakan 10 bins, berbeda dengan 20 bins saat mencari modus
ax.hist(my_dataset.Th, bins=10, density=True, edgecolor='k', label="Measurements Hist") 
# Menambahkan garis vertikal penanda nilai Modus (Modal Value), berwarna merah dengan ketebalan 2
ax.axvline(modal_value, color='red', label='Modal Value', linewidth=2)
# Menambahkan garis vertikal penanda nilai Median, berwarna ungu dengan ketebalan 2
ax.axvline(median, color='purple',  label='Median Value', linewidth=2)
# Menambahkan garis vertikal penanda nilai Rata-rata (Mean), berwarna pink dengan ketebalan 2
ax.axvline(a_mean, color='pink',  label='Arithmetic mean', linewidth=2)
ax.set_xlabel('Th [ppm]')                               # Memberi nama label pada sumbu X
ax.set_ylabel('Probability density')                    # Memberi nama label pada sumbu Y
ax.legend()                                             # Memunculkan kotak keterangan (legenda) agar pembaca tahu arti masing-masing warna garis
```

```python
import numpy as np                                      # Mengimpor pustaka NumPy untuk komputasi numerik matematis
from scipy.stats import skew                            # Mengimpor fungsi 'skew' khusus dari pustaka SciPy untuk menghitung kemiringan (skewness)

a_mean = my_dataset.Zr.mean()                           # Menghitung nilai rata-rata (mean) dari kolom 'Zr' (Zirconium)
median = my_dataset.Zr.median()                         # Menghitung nilai tengah (median) dari kolom 'Zr'
# Menghitung histogram dengan 20 kelompok (bins) untuk keperluan mencari puncak distribusi
hist, bin_edges = np.histogram(my_dataset['Zr'], bins=20, density=True)
# Mengestimasi nilai modus dengan mencari titik tengah dari bin yang frekuensi probabilitasnya paling tinggi
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1])/2
standard_deviation = my_dataset['Zr'].std()             # Menghitung nilai standar deviasi (simpangan baku) dari kolom 'Zr'

# Menghitung Koefisien Kemiringan Pearson Pertama (Pearson's first coefficient of skewness) menggunakan rumus: (Mean - Modus) / Standar Deviasi
a1 = (a_mean - modal_value) / standard_deviation 
# Menghitung Koefisien Kemiringan Pearson Kedua (Pearson's second coefficient of skewness) menggunakan rumus: 3 * (Mean - Median) / Standar Deviasi
a2 = 3 * (a_mean - median) / standard_deviation
# Menghitung Koefisien Kemiringan Fisher-Pearson langsung menggunakan fungsi bawaan dari SciPy
g1 = skew(my_dataset['Zr'])

print ('-------')                                       # Mencetak garis pemisah
# Mencetak hasil koefisien Pearson pertama dengan format 2 angka di belakang koma ({:.2f})
print ("Pearson's first coefficient of skewness: {:.2f}".format(a1))
# Mencetak hasil koefisien Pearson kedua dengan format 2 angka di belakang koma
print ("Pearson's 2nd moment of skewness: {:.2f}".format(a2))
# Mencetak hasil koefisien Fisher-Pearson dengan format 2 angka di belakang koma
print ("Fisher-Pearson's coefficient of skewness: {:.2f}".format(g1))
print ('-------')                                       # Mencetak garis pemisah penutup

'''
Output:
-------
Pearson's first coefficient of skewness: 0.74
Pearson's 2nd moment of skewness: 0.66
Fisher-Pearson's coefficient of skewness: 1.26
-------
'''
```
